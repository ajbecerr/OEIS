$a(n) = 24^{5n+2}$.

## Finding an exponential generating function:

Define:

$$EG(x)=\sum_{n=0}^{\infty}a(n)\frac{x^{n}}{n!}$$

For all $n\geq 0$, since:

$$a(n) = 24^{5n+2}$$

It follows that:

$$EG(x) = \sum_{n=0}^{\infty}24^{5n+2}\frac{x^{n}}{n!}$$

Or, equivalently:

$$EG(x) = 576\sum_{n=0}^{\infty}7962624^{n}\frac{x^{n}}{n!}$$

Therefore:

$$EG(x) = 576e^{7962624x}$$
