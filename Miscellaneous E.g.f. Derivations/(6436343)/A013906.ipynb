$a(n) = 23^{5n+1}$.

## Finding an exponential generating function:

Define:

$$EG(x)=\sum_{n=0}^{\infty}a(n)\frac{x^{n}}{n!}$$

For all $n\geq 0$, since:

$$a(n) = 23^{5n+1}$$

It follows that:

$$EG(x) = \sum_{n=0}^{\infty}23^{5n+1}\frac{x^{n}}{n!}$$

Or, equivalently:

$$EG(x) = 23\sum_{n=0}^{\infty}6436343\frac{x^{n}}{n!}$$

Therefore:

$$EG(x) = 23e^{6436343x}$$
