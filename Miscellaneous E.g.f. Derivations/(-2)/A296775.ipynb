Expansion of $\frac{\left(1-x\right)^{3}}{1+2x}$.

## Finding an exponential generating function:

Define:

$$EG(x)=\sum_{n=0}^{\infty}a(n)\frac{x^{n}}{n!}$$

Since:

$$a(0)=1$$

$$a(1)=-5$$

$$a(2)=13$$

And, for all $n\geq 3$:

$$a(n)=27\times\left(-1\right)^{n}\times2^{n-3}$$

It follows that:

$$EG(x)=1-5x+\frac{13}{2}x^{2}+\sum_{n=3}^{\infty}27\times\left(-1\right)^{n}\times2^{n-3}\frac{x^{n}}{n!}$$

Or, equivalently:

$$EG(x)=1-5x+\frac{13}{2}x^{2}+\frac{27}{8}\sum_{n=3}^{\infty}\left(-2\right)^{n}\frac{x^{n}}{n!}$$

Ergo:

$$EG(x)=1-5x+\frac{13}{2}x^{2}+\frac{27}{8}\left(e^{-2x}-2x^{2}+2x-1\right)$$

Therefore:

$$EG(x)=-\frac{19}{8}+\frac{7}{4}x-\frac{1}{4}x^{2}+\frac{27}{8}e^{-2x}$$
