## Finding an exponential generating function:

Define:

$$EG(x)=\sum_{n=0}^{\infty}a(n)\frac{x^{n}}{n!}$$

Since:

$$a(0)=5$$

And, for all $n\geq 1$:

$$a(n)=10$$

It follows that:

$$EG(x)=5+\sum_{n=1}^{\infty}10\frac{x^{n}}{n!}$$

Or, equivalently:

$$EG(x)=5+10\sum_{n=1}^{\infty}\frac{x^{n}}{n!}$$

Ergo:

$$EG(x)=5+10\left(e^{x}-1\right)$$

Therefore:

$$EG(x)=-5+10e^{x}$$
