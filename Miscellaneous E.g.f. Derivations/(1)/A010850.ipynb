Constant sequence: $a(n) = 11$.

## Finding an exponential generating function:

Define:

$$EG(x)=\sum_{n=0}^{\infty}a(n)\frac{x^{n}}{n!}$$

Since, for all $n\geq 0$:

$$a(n)=11$$

It follows that:

$$EG(x)=\sum_{n=0}^{\infty}11\frac{x^{n}}{n!}$$

Ergo:

$$EG(x)=11\sum_{n=0}^{\infty}\frac{x^{n}}{n!}$$

Therefore:

$$EG(x)=11e^{x}$$
