## Finding an exponential generating function:

Define:

$$EG(x)=\sum_{n=0}^{\infty}a(n)\frac{x^{n}}{n!}$$

Since:

$$a(1)=0$$

$$a(2)=1$$

And, for all $n\geq 3$:

$$a(n)=7$$

It follows that:

$$EG(x)=\frac{1}{2}x^{2}+\sum_{n=3}^{\infty}7\frac{x^{n}}{n!}$$

Or, equivalently:

$$EG(x)=\frac{1}{2}x^{2}+7\sum_{n=3}^{\infty}\frac{x^{n}}{n!}$$

Ergo:

$$EG(x)=\frac{1}{2}x^{2}+7\left(e^{x}-\frac{1}{2}x^{2}-x-1\right)$$

Therefore:

$$EG(x)=-7-7x-3x^{2}+7e^{x}$$
