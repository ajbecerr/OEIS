## Finding an exponential generating function:

Define:

$$EG(x)=\sum_{n=0}^{\infty}a(n)\frac{x^{n}}{n!}$$

Since:

$$a(0)=22$$

And, for all $n\geq 1$:

$$a(n)=44$$

It follows that:

$$EG(x)=22+\sum_{n=1}^{\infty}44\frac{x^{n}}{n!}$$

Or, equivalently:

$$EG(x)=22+44\sum_{n=1}^{\infty}\frac{x^{n}}{n!}$$

Ergo:

$$EG(x)=22+44\left(e^{x}-1\right)$$

Therefore:

$$EG(x)=-22+44e^{x}$$
