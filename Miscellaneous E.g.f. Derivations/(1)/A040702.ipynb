Continued fraction for $\sqrt{730}$.

## Finding an exponential generating function:

Define:

$$EG(x)=\sum_{n=0}^{\infty}a(n)\frac{x^{n}}{n!}$$

Since:

$$a(0)=27$$

And, for all $n\geq 1$:

$$a(n)=54$$

It follows that:

$$EG(x)=27+\sum_{n=1}^{\infty}54\frac{x^{n}}{n!}$$

Or, equivalently:

$$EG(x)=27+54\sum_{n=1}^{\infty}\frac{x^{n}}{n!}$$

Ergo:

$$EG(x)=27+54\left(e^{x}-1\right)$$

Therefore:

$$EG(x)=-27+54e^{x}$$
