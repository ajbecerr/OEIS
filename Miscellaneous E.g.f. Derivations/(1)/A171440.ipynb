Expansion of $\large \frac{\left(1+x\right)^{5}}{1-x}$.

## Finding an exponential generating function:

Define:

$$EG(x)=\sum_{n=0}^{\infty}a(n)\frac{x^{n}}{n!}$$

Since:

$$a(0)=1$$

$$a(1)=6$$

$$a(2)=16$$

$$a(3)=26$$

$$a(4)=31$$

And, for all $n\geq 5$:

$$a(n)=32$$

It follows that:

$$EG(x)=1+6x+8x^{2}+\frac{13}{3}x^{3}+\frac{31}{24}x^{4}+\sum_{n=5}^{\infty}32\frac{x^{n}}{n!}$$

Or, equivalently:

$$EG(x)=1+6x+8x^{2}+\frac{13}{3}x^{3}+\frac{31}{24}x^{4}+32\sum_{n=5}^{\infty}\frac{x^{n}}{n!}$$

Ergo:

$$EG(x)=1+6x+8x^{2}+\frac{13}{3}x^{3}+\frac{31}{24}x^{4}+32\left(e^{x}-\frac{1}{24}x^{4}-\frac{1}{6}x^{3}-\frac{1}{2}x^{2}-x-1\right)$$

Therefore:

$$EG(x)=-31-26x-8x^{2}-x^{3}-\frac{1}{24}x^{4}+32e^{x}$$
