Constant sequence: $a(n) = 23$.

## Finding an exponential generating function:

Define:

$$EG(x)=\sum_{n=0}^{\infty}a(n)\frac{x^{n}}{n!}$$

Since, for all $n\geq 0$:

$$a(n)=23$$

It follows that:

$$EG(x)=\sum_{n=0}^{\infty}23\frac{x^{n}}{n!}$$

Ergo:

$$EG(x)=23\sum_{n=0}^{\infty}\frac{x^{n}}{n!}$$

Therefore:

$$EG(x)=23e^{x}$$
