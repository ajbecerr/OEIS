Number of regular convex polytopes in $n$-dimensional space, or $-1$ if the number is infinite.

## Finding an exponential generating function:

Define:

$$EG(x)=\sum_{n=0}^{\infty}a(n)\frac{x^{n}}{n!}$$

Since:

$$a(0)=1$$

$$a(1)=1$$

$$a(2)=-1$$

$$a(3)=5$$

$$a(4)=6$$

And, for all $n\geq 5$:

$$a(n)=3$$

It follows that:

$$EG(x)=1+x-\frac{1}{2}x^{2}+\frac{5}{6}x^{3}+\frac{6}{24}x^{4}+\sum_{n=5}^{\infty}3\frac{x^{n}}{n!}$$

Or, equivalently:

$$EG(x)=1+x-\frac{1}{2}x^{2}+\frac{5}{6}x^{3}+\frac{6}{24}x^{4}+3\sum_{n=5}^{\infty}\frac{x^{n}}{n!}$$

Ergo:

$$EG(x)=1+x-\frac{1}{2}x^{2}+\frac{5}{6}x^{3}+\frac{6}{24}x^{4}+3\left(e^{x}-\frac{1}{24}x^{4}-\frac{1}{6}x^{3}-\frac{1}{2}x^{2}-x-1\right)$$

Therefore:

$$EG(x)=-2-2x-2x^{2}+\frac{1}{3}x^{3}+\frac{1}{8}x^{4}+3e^{x}$$
